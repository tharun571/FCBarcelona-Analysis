```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import numpy as np
```

```python
def shots():
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    
    import urllib.request, json
    with urllib.request.urlopen("https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/15946.json") as url:
        data = json.loads(url.read().decode())
        for data1 in data:
            ty=data1['type']
            if ty['name'] == 'Shot':
                team=data1['possession_team']
                
                if team['name']=='Barcelona':
                    sh=data1['shot']
                    xg=sh['statsbomb_xg']
                    out=sh['outcome']
                    x=data1['location'][0]
                    y=data1['location'][1]
                    
                    if out['name'] == "Goal":
                        shot=plt.Circle((x,y),np.sqrt(xg*15),color='blue')
                    else:
                        shot=plt.Circle((x,y),np.sqrt(xg*15),color='deepskyblue')
                        
                else:
                    sh=data1['shot']
                    xg=sh['statsbomb_xg']
                    out=sh['outcome']
                    x=data1['location'][0]
                    y=data1['location'][1]
                    if out['name'] == "Goal":
                        shot=plt.Circle((130-x,y),np.sqrt(xg*15),color='green')
                    else:
                        shot=plt.Circle((130-x,y),np.sqrt(xg*15),color='lightgreen')
                    
                ax.add_patch(shot)
    
    #Pitch Outline & Centre Line
    plt.plot([0,0],[0,90], color="black")
    plt.plot([0,130],[90,90], color="black")
    plt.plot([130,130],[90,0], color="black")
    plt.plot([130,0],[0,0], color="black")
    plt.plot([65,65],[0,90], color="black")
    
    #Left Penalty Area
    plt.plot([16.5,16.5],[65,25],color="black")
    plt.plot([0,16.5],[65,65],color="black")
    plt.plot([16.5,0],[25,25],color="black")
    
    #Right Penalty Area
    plt.plot([130,113.5],[65,65],color="black")
    plt.plot([113.5,113.5],[65,25],color="black")
    plt.plot([113.5,130],[25,25],color="black")
    
    #Left 6-yard Box
    plt.plot([0,5.5],[54,54],color="black")
    plt.plot([5.5,5.5],[54,36],color="black")
    plt.plot([5.5,0.5],[36,36],color="black")
    
    #Right 6-yard Box
    plt.plot([130,124.5],[54,54],color="black")
    plt.plot([124.5,124.5],[54,36],color="black")
    plt.plot([124.5,130],[36,36],color="black")
    
    #Prepare Circles
    centreCircle = plt.Circle((65,45),9.15,color="black",fill=False)
    centreSpot = plt.Circle((65,45),0.8,color="black")
    leftPenSpot = plt.Circle((11,45),0.8,color="black")
    rightPenSpot = plt.Circle((119,45),0.8,color="black")
    
    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)
    
    #Prepare Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="black")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
    
    #Tidy Axes
    plt.text(150,45,"Shots taken by FC Barcelona and Deportivo Alaves")
    
    c1=plt.Circle((150,40),1,color="blue")
    plt.text(152,40," - Goals by Barcelona")
    c2=plt.Circle((150,35),1,color="deepskyblue")
    plt.text(152,35," - Shots by Barcelona")
    
    cc=plt.Circle((200,45),0.1,color="white")
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.add_patch(cc)
    
    c11=plt.Circle((150,30),1,color="green")
    plt.text(152,30," - Goals by Deportivo Alaves")
    c21=plt.Circle((150,25),1,color="lightgreen")
    plt.text(152,25," - Shots by Deportivo Alaves")
    
    ax.add_patch(c11)
    ax.add_patch(c21)
    
    
    fig.set_size_inches(15,7)
    
    
    #Display Pitch
    plt.show()
  
```
