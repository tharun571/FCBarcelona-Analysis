```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import numpy as np
import math
from matplotlib.patches import Arc
import seaborn as sns
```

```python
shots1=np.empty((0,2),float)
goals1=np.empty((0,2),float)
xg1=np.array([],dtype=float)

shots2=np.empty((0,2),float)
goals2=np.empty((0,2),float)
xg2=np.array([],dtype=float)

shots3=np.empty((0,2),float)
goals3=np.empty((0,2),float)
xg3=np.array([],dtype=float)

shots4=np.empty((0,2),float)
goals4=np.empty((0,2),float)
xg4=np.array([],dtype=float)
```

```python
listt=["1","2","4","21","22","23","24","25","26","27","37","38","39","40","41"]
list1=np.array([])
list2=np.array([])
list3=np.array([])
list4=np.array([])

for a in listt:
    import urllib.request, json
    with urllib.request.urlopen("https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/11/"+a+".json") as url:
         data = json.loads(url.read().decode())
         for data1 in data:
            n=data1['season']['season_name']
            if n == "2007/2008" or n== "2008/2009" or n== "2009/2010":
                list1=np.append(list1,str(data1['match_id']))
            elif n == "2010/2011" or n== "2011/2012" or n== "2012/2013":
                list2=np.append(list2,str(data1['match_id']))
            elif n == "2013/2014" or n== "2014/2015" or n== "2015/2016":
                list3=np.append(list3,str(data1['match_id']))
            elif n == "2016/2017" or n== "2017/2018" or n== "2018/2019":
                list4=np.append(list4,str(data1['match_id'])) 
                
print(list1.shape)
print(list2.shape)
print(list3.shape)
print(list4.shape)         
```

```python
for a in list1:
        import urllib.request, json
        with urllib.request.urlopen("https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/"+a+".json") as url:
         data = json.loads(url.read().decode())
        for data1 in data:
            ty=data1['type']
            if ty['name'] == 'Shot':
                team=data1['team']
                if team['name']=='Barcelona':
                    if data1['player']['name'] == 'Lionel Andrés Messi Cuccittini':
                        sh=data1['shot']
                        xg=sh['statsbomb_xg']
                        out=sh['outcome']
                        x=data1['location'][0]
                        y=data1['location'][1]
                        shots1=np.append(shots1,np.array([[x,y]]),axis=0)
                        if out['name'] == "Goal":
                            goals1=np.append(goals1,np.array([[x,y]]),axis=0)
                            xg1=np.append(xg1,data1['shot']['statsbomb_xg'])
                            
                            
                            
```
