```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import math
import seaborn as sns
```

```
#Getting the match ids of the matches in that particular season
ids = ['1','2','4','21','22','23','24','25','26','27','37','38','39','40','41','42']
match1=np.array([])
match2=np.array([])
match3=np.array([])
for id in ids:
        import urllib.request, json
        with urllib.request.urlopen("https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/11/"+id+".json") as url:
          data = json.loads(url.read().decode())
        for data1 in data:
          if data1['season']['season_name'] == '2019/2020':
            match1 = np.append(match1,str(data1['match_id']))
          elif data1['season']['season_name'] == '2014/2015':
            match2 = np.append(match2,str(data1['match_id']))  
          if data1['season']['season_name'] == '2009/2010':
            match3 = np.append(match3,str(data1['match_id']))

print(match1.shape) 
print(match2.shape) 
print(match3.shape) 
```

```
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    count1 =0
    xG1=0

    for a in match1:
        count1 = count1 +1
        import urllib.request, json
        with urllib.request.urlopen("https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/"+a+".json") as url:
         data = json.loads(url.read().decode())
        for data1 in data:
            if data1['type']['name'] == 'Shot' :
              if data1['possession_team']['name'] == 'Barcelona':
                xG1 = xG1 + data1['shot']['statsbomb_xg']
              if data1['shot']['outcome']['name'] == 'Goal' :
                x = data1['location'][0]
                y = data1['location'][1]
                if data1['possession_team']['name'] == 'Barcelona':
                  
                  shotCircle = plt.Circle((x,y),.75,color = 'blue')  
                else:
                  shotCircle = plt.Circle((130-x,90-y),.75,color = 'red')
                ax.add_patch(shotCircle)


   
    plt.plot([0,0],[0,90], color="white")
    plt.plot([0,130],[90,90], color="white")
    plt.plot([130,130],[90,0], color="white")
    plt.plot([130,0],[0,0], color="white")
    plt.plot([65,65],[0,90], color="white")
    
    #Left Penalty Area
    plt.plot([16.5,16.5],[65,25],color="white")
    plt.plot([0,16.5],[65,65],color="white")
    plt.plot([16.5,0],[25,25],color="white")
    
    #Right Penalty Area
    plt.plot([130,113.5],[65,65],color="white")
    plt.plot([113.5,113.5],[65,25],color="white")
    plt.plot([113.5,130],[25,25],color="white")
    
    #Left 6-yard Box
    plt.plot([0,5.5],[54,54],color="white")
    plt.plot([5.5,5.5],[54,36],color="white")
    plt.plot([5.5,0.5],[36,36],color="white")
    
    #Right 6-yard Box
    plt.plot([130,124.5],[54,54],color="white")
    plt.plot([124.5,124.5],[54,36],color="white")
    plt.plot([124.5,130],[36,36],color="white")
    
    #Prepare Circles
    centreCircle = plt.Circle((65,45),9.15,color="white",fill=False)
    centreSpot = plt.Circle((65,45),0.8,color="white")
    leftPenSpot = plt.Circle((11,45),0.8,color="white")
    rightPenSpot = plt.Circle((119,45),0.8,color="white")
    
    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)

    ax.add_patch(shotCircle)
    
    #Prepare Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="white")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="white")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)

    fig.set_size_inches(10,7)
    
    plt.ylim(0, 90)
    plt.xlim(0, 130)
    
    plt.style.use('dark_background')
    ax.set_title("Goals in 19/20 Season in 33 matches")
    #Display Pitch
    plt.show()                
```
